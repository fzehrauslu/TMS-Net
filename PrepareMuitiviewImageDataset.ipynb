```python
def FolderPath(FolderNo,IndiceOfFile):
    
    pathToSave='PathToDataset/'+str(FolderNo)
    TrainingPath=os.path.join(pathToSave,'training')
    ValidationPath=os.path.join(pathToSave,'validation')
    TestPath=os.path.join(pathToSave,'test')

    if IndiceOfFile>79 and IndiceOfFile<100 :
        ## save data to test set  
        PathToIm=os.path.join(TestPath,'images')
        PathToGT=os.path.join(TestPath,'1st_manual')

    elif IndiceOfFile>-1 and IndiceOfFile<10 :
    ## save data to validation set 
        PathToIm=os.path.join(ValidationPath,'images')
        PathToGT=os.path.join(ValidationPath,'1st_manual')        
    else:
        ## save data to training set
        PathToIm=os.path.join(TrainingPath,'images')
        PathToGT=os.path.join(TrainingPath,'1st_manual')

    return PathToIm, PathToGT
```

```python
def SaveImagesToFolders(files,IndiceOfFile,WholeVolume=False,CreateTrainingValImages=True,TestNumpyFiles=True):  
            
         
            
    ED,ED_GT=Load_Im_Endo(files)


    print(IndiceOfFileOri)        

        

    if CreateTrainingValImages:  
        if WholeVolume:
            StartSlice,EndSlice=0,ED_GT.shape[2]
            EndSlice=EndSlice-1
        else:
            StartSlice,EndSlice= FindSlicesWithLabels(ED_GT>0.2)
            

        if WholeVolume:
           
            ## slices 3d images along specific axes
            # FolderNo=9 contains all sliced images from the three views
            # FolderNo=8 contains all sliced images from the sagittal view
            # FolderNo=7 contains all sliced images from the coronal view
            # FolderNo=6 contains all sliced images from the axial view     
            
            # test data is saved to FolderNo=8
            
            StartSlice,EndSlice= FindSlicesWithLabels(ED_GT.permute(1,2,0)>0.2)
            
            for i in range(0, StartSlice-1,5):
                FolderNo=8
                PathToIm, PathToGT=FolderPath(FolderNo,IndiceOfFile)  

                torchvision.utils.save_image(ED.permute(1,2,0)[:,:,i],os.path.join(PathToIm,'Slice_S_'+str(IndiceOfFileOri)+'_'+str(i)+'.png'),normalize=False)
                torchvision.utils.save_image(ED_GT.permute(1,2,0)[:,:,i],os.path.join(PathToGT,'Slice_S'+str(IndiceOfFileOri)+'_'+str(i)+'.png'),normalize=False)
                
                FolderNo=9
                PathToIm, PathToGT=FolderPath(FolderNo,IndiceOfFile)  

                torchvision.utils.save_image(ED.permute(1,2,0)[:,:,i],os.path.join(PathToIm,'Slice_S_'+str(IndiceOfFileOri)+'_'+str(i)+'.png'),normalize=False)
                torchvision.utils.save_image(ED_GT.permute(1,2,0)[:,:,i],os.path.join(PathToGT,'Slice_S'+str(IndiceOfFileOri)+'_'+str(i)+'.png'),normalize=False)


                
            
            for i in range(StartSlice-1,EndSlice+1):
                FolderNo=8
                PathToIm, PathToGT=FolderPath(FolderNo,IndiceOfFile)  

                torchvision.utils.save_image(ED.permute(1,2,0)[:,:,i],os.path.join(PathToIm,'Slice_S_'+str(IndiceOfFileOri)+'_'+str(i)+'.png'),normalize=False)
                torchvision.utils.save_image(ED_GT.permute(1,2,0)[:,:,i],os.path.join(PathToGT,'Slice_S'+str(IndiceOfFileOri)+'_'+str(i)+'.png'),normalize=False)
                
                FolderNo=9
                PathToIm, PathToGT=FolderPath(FolderNo,IndiceOfFile)  

                torchvision.utils.save_image(ED.permute(1,2,0)[:,:,i],os.path.join(PathToIm,'Slice_S_'+str(IndiceOfFileOri)+'_'+str(i)+'.png'),normalize=False)
                torchvision.utils.save_image(ED_GT.permute(1,2,0)[:,:,i],os.path.join(PathToGT,'Slice_S'+str(IndiceOfFileOri)+'_'+str(i)+'.png'),normalize=False)

            for i in range(EndSlice+1,ED_GT.permute(1,2,0).shape[2],5):
                FolderNo=8
                PathToIm, PathToGT=FolderPath(FolderNo,IndiceOfFile)  

                torchvision.utils.save_image(ED.permute(1,2,0)[:,:,i],os.path.join(PathToIm,'Slice_S_'+str(IndiceOfFileOri)+'_'+str(i)+'.png'),normalize=False)
                torchvision.utils.save_image(ED_GT.permute(1,2,0)[:,:,i],os.path.join(PathToGT,'Slice_S'+str(IndiceOfFileOri)+'_'+str(i)+'.png'),normalize=False)
                
                FolderNo=9
                PathToIm, PathToGT=FolderPath(FolderNo,IndiceOfFile)  

                torchvision.utils.save_image(ED.permute(1,2,0)[:,:,i],os.path.join(PathToIm,'Slice_S_'+str(IndiceOfFileOri)+'_'+str(i)+'.png'),normalize=False)
                torchvision.utils.save_image(ED_GT.permute(1,2,0)[:,:,i],os.path.join(PathToGT,'Slice_S'+str(IndiceOfFileOri)+'_'+str(i)+'.png'),normalize=False)


                
                
            ######################################
            StartSlice,EndSlice= FindSlicesWithLabels(ED_GT.permute(0,1,2)>0.2)
                                                      
            for i in range(0,StartSlice-1,5):
                FolderNo=7
                PathToIm, PathToGT=FolderPath(FolderNo,IndiceOfFile)  

                torchvision.utils.save_image(ED.permute(0,1,2)[:,:,i],os.path.join(PathToIm,'Slice_C_'+str(IndiceOfFileOri)+'_'+str(i)+'.png'),normalize=False)
                torchvision.utils.save_image(ED_GT.permute(0,1,2)[:,:,i],os.path.join(PathToGT,'Slice_C'+str(IndiceOfFileOri)+'_'+str(i)+'.png'),normalize=False)

                FolderNo=9
                PathToIm, PathToGT=FolderPath(FolderNo,IndiceOfFile)  

                torchvision.utils.save_image(ED.permute(0,1,2)[:,:,i],os.path.join(PathToIm,'Slice_C_'+str(IndiceOfFileOri)+'_'+str(i)+'.png'),normalize=False)
                torchvision.utils.save_image(ED_GT.permute(0,1,2)[:,:,i],os.path.join(PathToGT,'Slice_C'+str(IndiceOfFileOri)+'_'+str(i)+'.png'),normalize=False)


                                                      

            for i in range(StartSlice-1,EndSlice+1):
                FolderNo=7
                PathToIm, PathToGT=FolderPath(FolderNo,IndiceOfFile)  

                torchvision.utils.save_image(ED.permute(0,1,2)[:,:,i],os.path.join(PathToIm,'Slice_C_'+str(IndiceOfFileOri)+'_'+str(i)+'.png'),normalize=False)
                torchvision.utils.save_image(ED_GT.permute(0,1,2)[:,:,i],os.path.join(PathToGT,'Slice_C'+str(IndiceOfFileOri)+'_'+str(i)+'.png'),normalize=False)

                FolderNo=9
                PathToIm, PathToGT=FolderPath(FolderNo,IndiceOfFile)  

                torchvision.utils.save_image(ED.permute(0,1,2)[:,:,i],os.path.join(PathToIm,'Slice_C_'+str(IndiceOfFileOri)+'_'+str(i)+'.png'),normalize=False)
                torchvision.utils.save_image(ED_GT.permute(0,1,2)[:,:,i],os.path.join(PathToGT,'Slice_C'+str(IndiceOfFileOri)+'_'+str(i)+'.png'),normalize=False)

            for i in range(EndSlice+1,ED.permute(0,1,2).shape[2],5):
                FolderNo=7
                PathToIm, PathToGT=FolderPath(FolderNo,IndiceOfFile)  

                torchvision.utils.save_image(ED.permute(0,1,2)[:,:,i],os.path.join(PathToIm,'Slice_C_'+str(IndiceOfFileOri)+'_'+str(i)+'.png'),normalize=False)
                torchvision.utils.save_image(ED_GT.permute(0,1,2)[:,:,i],os.path.join(PathToGT,'Slice_C'+str(IndiceOfFileOri)+'_'+str(i)+'.png'),normalize=False)

                FolderNo=9
                PathToIm, PathToGT=FolderPath(FolderNo,IndiceOfFile)  

                torchvision.utils.save_image(ED.permute(0,1,2)[:,:,i],os.path.join(PathToIm,'Slice_C_'+str(IndiceOfFileOri)+'_'+str(i)+'.png'),normalize=False)
                torchvision.utils.save_image(ED_GT.permute(0,1,2)[:,:,i],os.path.join(PathToGT,'Slice_C'+str(IndiceOfFileOri)+'_'+str(i)+'.png'),normalize=False)


                                                      
            ######################################                                          
            StartSlice,EndSlice= FindSlicesWithLabels(ED_GT.permute(0,2,1)>0.2)                                          
            for i in range(0,StartSlice-1,5):
                FolderNo=6
                PathToIm, PathToGT=FolderPath(FolderNo,IndiceOfFile)  

        
                torchvision.utils.save_image(ED.permute(0,2,1)[:,:,i],os.path.join(PathToIm,'Slice_A_'+str(IndiceOfFileOri)+'_'+str(i)+'.png'),normalize=False)
                torchvision.utils.save_image(ED_GT.permute(0,2,1)[:,:,i],os.path.join(PathToGT,'Slice_A'+str(IndiceOfFileOri)+'_'+str(i)+'.png'),normalize=False)

                FolderNo=9
                PathToIm, PathToGT=FolderPath(FolderNo,IndiceOfFile)  

        
                torchvision.utils.save_image(ED.permute(0,2,1)[:,:,i],os.path.join(PathToIm,'Slice_A_'+str(IndiceOfFileOri)+'_'+str(i)+'.png'),normalize=False)
                torchvision.utils.save_image(ED_GT.permute(0,2,1)[:,:,i],os.path.join(PathToGT,'Slice_A'+str(IndiceOfFileOri)+'_'+str(i)+'.png'),normalize=False)


            
            for i in range(StartSlice-1,EndSlice+1):
                FolderNo=6
                PathToIm, PathToGT=FolderPath(FolderNo,IndiceOfFile)  

        
                torchvision.utils.save_image(ED.permute(0,2,1)[:,:,i],os.path.join(PathToIm,'Slice_A_'+str(IndiceOfFileOri)+'_'+str(i)+'.png'),normalize=False)
                torchvision.utils.save_image(ED_GT.permute(0,2,1)[:,:,i],os.path.join(PathToGT,'Slice_A'+str(IndiceOfFileOri)+'_'+str(i)+'.png'),normalize=False)

                FolderNo=9
                PathToIm, PathToGT=FolderPath(FolderNo,IndiceOfFile)  

        
                torchvision.utils.save_image(ED.permute(0,2,1)[:,:,i],os.path.join(PathToIm,'Slice_A_'+str(IndiceOfFileOri)+'_'+str(i)+'.png'),normalize=False)
                torchvision.utils.save_image(ED_GT.permute(0,2,1)[:,:,i],os.path.join(PathToGT,'Slice_A'+str(IndiceOfFileOri)+'_'+str(i)+'.png'),normalize=False)

            
            
            
            
            for i in range(EndSlice+1,ED.permute(0,2,1).shape[2],5):
                FolderNo=6
                PathToIm, PathToGT=FolderPath(FolderNo,IndiceOfFile)  

        
                torchvision.utils.save_image(ED.permute(0,2,1)[:,:,i],os.path.join(PathToIm,'Slice_A_'+str(IndiceOfFileOri)+'_'+str(i)+'.png'),normalize=False)
                torchvision.utils.save_image(ED_GT.permute(0,2,1)[:,:,i],os.path.join(PathToGT,'Slice_A'+str(IndiceOfFileOri)+'_'+str(i)+'.png'),normalize=False)

                FolderNo=9
                PathToIm, PathToGT=FolderPath(FolderNo,IndiceOfFile)  

        
                torchvision.utils.save_image(ED.permute(0,2,1)[:,:,i],os.path.join(PathToIm,'Slice_A_'+str(IndiceOfFileOri)+'_'+str(i)+'.png'),normalize=False)
                torchvision.utils.save_image(ED_GT.permute(0,2,1)[:,:,i],os.path.join(PathToGT,'Slice_A'+str(IndiceOfFileOri)+'_'+str(i)+'.png'),normalize=False)

           
 

    if TestNumpyFiles:
            #pdb.set_trace()
            if len(ED_GT)>0:
                if WholeVolume:
                    StartSlice,EndSlice=0,ED_GT.shape[2]
                    EndSlice=EndSlice-1
                else:
                    StartSlice,EndSlice= FindSlicesWithLabels(ED_GT>0.2)
                

                if WholeVolume:
                    FolderNo=8
                    PathToIm, PathToGT=FolderPath(FolderNo,IndiceOfFile)  
                    
                    np.save(os.path.join(PathToIm,'Numpy_ED_'+str(IndiceOfFileOri)+'.npy'), ED)
                    np.save(os.path.join(PathToGT,'Numpy_ED_GT_'+str(IndiceOfFileOri)+'.npy'), ED_GT)
                    

 
            


def FindSlicesWithLabels(ED_GT):
    ## Width, height, depth
    SumOfLabels= torch.sum(ED_GT.view(-1,ED_GT.shape[2]),0)
    if len(np.argwhere(SumOfLabels>0))==0:
        pdb.set_trace()
    print(np.argwhere(SumOfLabels>0))
    StartSlice= torch.min(np.argwhere(SumOfLabels>0))
    EndSlice= torch.max(np.argwhere(SumOfLabels>0))
    print(StartSlice,EndSlice,ED_GT.shape[2])
    return StartSlice,EndSlice


def Load_Im_Endo(files):
    ## load ED and ES data and their ground truth images
    #print(np.shape(files)[0])
    #pdb.set_trace()
    Data4D=[]
    ED_GT=[]
    ES_GT=[]
    AS_GT=[]
    
    ED=[]
    ES=[]
    AS=[]
    
    

    FileNames=['/lgemri.nrrd','/laendo.nrrd']
    
    FileNames_V=[ED,ED_GT]
    #pdb.set_trace()
    for j in range(len(FileNames)):
        for i in range(np.shape(files)[0]):
            if files[i].endswith(FileNames[j]):

                FileNames_V[j],header=nrrd.read(files[i])

                
                
                FileNames_V[j]=torch.from_numpy(np.asarray(FileNames_V[j]))
                FileNames_V[j]= MakeTensorIsotropic2(FileNames_V[j],[0.5,0.5,1],interpolationMet='nearest')
                print(FileNames_V[j].shape)

    
    if len(FileNames_V[0])>0:
        FileNames_V[0],FileNames_V[1]= NormaliseIntensity(FileNames_V[0],FileNames_V[1],True)

    
    return FileNames_V[0],FileNames_V[1]

 
def MakeTensorIsotropic2(Im,zoom,interpolationMet):
    z0=zoom[0]
    z1=zoom[1]
    z2=zoom[2]
    
    Im=Im[None,None,:,:,:].type(torch.FloatTensor)
    
    m=nn.functional.interpolate(Im,scale_factor=(z0,z1,z2), mode=interpolationMet)

    return m[0,0,:,:,:]
def MakeFixedSizeOfImages(ED):
    w,c,d=ED.shape
    if w>c:
        maxV=w
    elif c>w:
        maxV=c
    else:
        maxV=w
        
        
    if maxV%16==0:
        size= maxV
        
    else:
        size=maxV+16-maxV%16
           
    EDnew=torch.zeros(size,size,d)
    difW=(size-w)/2
    difC=(size-c)/2
    EDnew[difW:difW+w,difC:difC+c,:]=ED
    return EDnew    
    
def NormaliseIntensity(ES,ES_GT,RemoveOutliers=False):
    #pdb.set_trace()
    if RemoveOutliers:
    
        Quantile1=np.percentile(ES, 2) #2
        Quantile99=np.percentile(ES, 99) # 99

        Quantile1_Loc=np.where(ES < Quantile1)
        Quantile99_Loc=np.where(ES > Quantile99)
        #pdb.set_trace()
        if len(Quantile1_Loc)>0:
            x=Quantile1_Loc[0]
            y=Quantile1_Loc[1]
            z=Quantile1_Loc[2]

            ES[x,y,z]=Quantile1

        if len(Quantile99_Loc)>0:
            x=Quantile99_Loc[0]
            y=Quantile99_Loc[1]
            z=Quantile99_Loc[2]
            ES[x,y,z]=Quantile99    
    
        
    ES=ES.type(torch.FloatTensor)
    ES_GT=ES_GT.type(torch.FloatTensor)
    
    ES=MakeFixedSizeOfImages(ES)
    ES_GT=MakeFixedSizeOfImages(ES_GT)
    
    w,c,d=np.shape(ES)
    ES_2D=np.reshape(ES,(w*c,d))
    ES=(ES-torch.min(ES_2D,0)[0])/(torch.max(ES_2D,0)[0]-torch.min(ES_2D,0)[0])
    

    
    return ES,ES_GT

def FindSlicesWithLabels(ED_GT):
    SumOfLabels= torch.sum(ED_GT.view(-1,ED_GT.shape[2]),0)
    if len(np.argwhere(SumOfLabels>0))==0:
        pdb.set_trace()
    #print(np.argwhere(SumOfLabels>0))
    StartSlice= torch.min(np.argwhere(SumOfLabels>0))
    EndSlice= torch.max(np.argwhere(SumOfLabels>0))
    print(StartSlice,EndSlice,ED_GT.shape[2])
    return StartSlice,EndSlice
def StdBetweenSlices(Im1,Im2,Im3):
    std12=np.std(np.concatenate((Im1[:,:,None],Im2[:,:,None]),2),2)
    std23=np.std(np.concatenate((Im3[:,:,None],Im2[:,:,None]),2),2)
    return std12,std23 
```
